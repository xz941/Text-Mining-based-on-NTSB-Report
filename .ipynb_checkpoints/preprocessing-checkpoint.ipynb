```python
import pandas as pd
import numpy as np
```

```python
#Load data
mypath = "data/NTSB"
Occurrences = pd.read_csv(mypath+'/Occurrences.csv')
sequence = pd.read_csv(mypath+'/seq_of_events.csv')
aircraft = pd.read_csv(mypath+'/aircraft.csv')
events = pd.read_csv(mypath+'/events.csv')
dataDict = pd.read_csv(mypath+'/eADMSPUB_DataDictionary.csv')
# select key columns
events_select = events[['ev_id','ev_highest_injury','ev_year']]
aircraft_select = aircraft[['ev_id', 'Aircraft_Key', 'damage']]
Occurrences_select = Occurrences[['ev_id', 'Aircraft_Key', 'Occurrence_No', 'Occurrence_Code',
       'Phase_of_Flight']]
sequence_select = sequence[['ev_id', 'Aircraft_Key', 'Occurrence_No','seq_event_no','Subj_Code']]
# Merge tables according to event id, occurrence number, aircraft id
merge_buffer = pd.merge(Occurrences_select, sequence_select, on=['ev_id', 'Aircraft_Key', 'Occurrence_No'])
merge_buffer = pd.merge(merge_buffer, aircraft_select, on=['ev_id', 'Aircraft_Key'])
data = pd.merge(merge_buffer, events_select, on=['ev_id'])
data.head(10)
```

```python
#WITH '000' and '999'
data_event = data.dropna().copy()
occurrence_code_corpus = ['000','999'] + [str(x) for x in set(data_event['Occurrence_Code'])] +['']
subject_code_corpus = ['000','999'] + [str(x) for x in set(data_event['Subj_Code'])] +['']
phase_code_corpus = ['000','999'] + [str(x) for x in set(data_event['Phase_of_Flight'])] +['']
occurrence_list = data_event.loc[:,'Occurrence_Code'].astype(str)
phase_list = data_event.loc[:,'Phase_of_Flight'].astype(str)
subject_list = data_event.loc[:,'Subj_Code'].astype(str)

data_event.loc[:,'Phase_of_Flight'] = [phase_code_corpus.index(x) for x in phase_list]
data_event.loc[:,'Occurrence_Code'] = [occurrence_code_corpus.index(x) for x in occurrence_list]
data_event.loc[:,'Subj_Code'] = [subject_code_corpus.index(x) for x in subject_list]
data_group = data_event.groupby(['ev_id','Aircraft_Key'])[['Occurrence_No','Occurrence_Code','Phase_of_Flight','Subj_Code','damage','ev_highest_injury']].agg(lambda x: x.tolist()).reset_index()

data_group['damage'] = [x[0] for x in data_group['damage']]
data_group['ev_highest_injury'] = [x[0] for x in data_group['ev_highest_injury']]
data_group = data_group[data_group['damage'] != 'UNK ']
data_group = data_group[data_group['damage'] != 'NONE']
data_group.loc[data_group['damage'] == 'SUBS', 'damage'] = 1
data_group.loc[data_group['damage'] == 'MINR', 'damage'] = 0
data_group.loc[data_group['damage'] == 'DEST', 'damage'] = 2
data_group.head(3)
```

```python
for x in data_group.index:
    data_group.at[x,'Occurrence_Code'] = [0] + data_group.loc[x,'Occurrence_Code'] + [1]
    data_group.at[x,'Phase_of_Flight'] = [0] + data_group.loc[x,'Phase_of_Flight'] + [1]
    data_group.at[x,'Subj_Code'] = [0] + data_group.loc[x,'Subj_Code'] + [1]


buffer = data_event.drop_duplicates(["Occurrence_Code", "Subj_Code"])
occurrence_to_subj = buffer.groupby(['Occurrence_Code'])['Subj_Code'].agg(lambda x: x.tolist()).reset_index()
occurrence_to_subj['len'] = [len(x) for x in occurrence_to_subj['Subj_Code']]
occurrence_to_subj = pd.concat([pd.DataFrame({'Occurrence_Code':[1],'Subj_Code':[1],'len':[1]}), occurrence_to_subj]).reset_index(drop = True) 
occurrence_to_subj = pd.concat([pd.DataFrame({'Occurrence_Code':[0],'Subj_Code':[0],'len':[1]}), occurrence_to_subj]).reset_index(drop = True)


buffer = data_event.drop_duplicates(["Occurrence_Code", "Subj_Code"])
subj_to_occurrence = buffer.groupby(['Subj_Code'])['Occurrence_Code'].agg(lambda x: x.tolist()).reset_index()
subj_to_occurrence['len'] = [len(x) for x in subj_to_occurrence['Occurrence_Code']]
subj_to_occurrence = pd.concat([pd.DataFrame({'Subj_Code':[1],'Occurrence_Code':[1],'len':[1]}), subj_to_occurrence]).reset_index(drop = True) 
subj_to_occurrence = pd.concat([pd.DataFrame({'Subj_Code':[0],'Occurrence_Code':[0],'len':[1]}), subj_to_occurrence]).reset_index(drop = True)


buffer = data_event.drop_duplicates(["Phase_of_Flight", "Occurrence_Code"])
phase_to_occurrence = buffer.groupby(['Phase_of_Flight'])['Occurrence_Code'].agg(lambda x: x.tolist()).reset_index()
phase_to_occurrence['len'] = [len(x) for x in phase_to_occurrence['Occurrence_Code']]
phase_to_occurrence = pd.concat([pd.DataFrame({'Phase_of_Flight':[1],'Occurrence_Code':[1],'len':[1]}), phase_to_occurrence]).reset_index(drop = True) 
phase_to_occurrence = pd.concat([pd.DataFrame({'Phase_of_Flight':[0],'Occurrence_Code':[0],'len':[1]}), phase_to_occurrence]).reset_index(drop = True)


buffer = data_event.drop_duplicates(["Phase_of_Flight", "Occurrence_Code"])
occurrence_to_phase = buffer.groupby(['Occurrence_Code'])['Phase_of_Flight'].agg(lambda x: x.tolist()).reset_index()
occurrence_to_phase['len'] = [len(x) for x in occurrence_to_phase['Phase_of_Flight']]
occurrence_to_phase = pd.concat([pd.DataFrame({'Occurrence_Code':[1],'Phase_of_Flight':[1],'len':[1]}), occurrence_to_phase]).reset_index(drop = True) 
occurrence_to_phase = pd.concat([pd.DataFrame({'Occurrence_Code':[0],'Phase_of_Flight':[0],'len':[1]}), occurrence_to_phase]).reset_index(drop = True)
```

```python
data_group.to_pickle("data/data_group.pkl")
occurrence_to_subj.to_pickle("data/occurrence_to_subj.pkl")
phase_to_occurrence.to_pickle("data/phase_to_occurrence.pkl")
data_event.to_pickle("data/data_event.pkl")
subj_to_occurrence.to_pickle("data/subj_to_occurrence.pkl")
occurrence_to_phase.to_pickle("data/occurrence_to_phase.pkl")
np.save("data/occurrence_code_corpus", np.array(occurrence_code_corpus))
np.save("data/subject_code_corpus", np.array(subject_code_corpus))
np.save("data/phase_code_corpus", np.array(phase_code_corpus))
```
